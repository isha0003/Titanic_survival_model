```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv("/Users/ishashree/Documents/titanic.csv")
df.sample(20)
```

```python
ss=df[["Survived","Pclass","Sex","Age","Fare"]]
ss
```

```python
m=ss["Age"].mean()
```

```python
m
```

```python
d=ss.fillna(m)
```

```python
d
```

```python
d
```

```python
from sklearn.preprocessing import LabelEncoder
Sex_n=LabelEncoder()
```

```python
Sex_n
```

```python
d["sex_n"]=Sex_n.fit_transform(d["Sex"])
```

```python
d
```

```python
final=d.drop(['Sex'],axis="columns")
final
```

```python
dependant=final.drop(["Survived"],axis="columns")
target=final["Survived"]
```

```python
dependant
```

```python
target
```

```python
from sklearn import tree
dt=tree.DecisionTreeClassifier()
```

```python
dt.fit(dependant,target)
```

```python
dt.predict([[1,26,30,1]])
```

```python
dt.predict([[3,45,12,1]])
```

```python
dt.score(dependant,target)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(dependant,target,test_size=0.2)
```

```python
dt.fit(x_train,y_train)
```

```python
dt.predict(x_test)
```

```python
dt.score(x_test,y_test)
```

```python
y_pred=dt.predict(x_test)
```

```python
cf=confusion_matrix(y_test,y_pred)
```

```python
cf
```

```python
import seaborn as sn

sn.heatmap(cf,annot=True)
plt.xlabel("Actual")
plt.ylabel("Predicted")

plt.show
```
